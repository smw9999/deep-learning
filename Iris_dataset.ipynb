## 1. 데이터 정의
---

* sklearn 으로 부터 iris 데이터셋을 불러옵니다.
* x_data와 y_data를 정의합니다.

```python
import pandas as pd

data=pd.read_csv("iris_train.csv")
data
```

```python
data_np=data.to_numpy()

x_data= data_np[:,2:].astype(float)
y_data=data_np[:,1]

y_data_label=[]

for y in y_data:
  if y=='setosa':
    y_data_label.append(0)
  elif y=='versicolor':
    y_data_label.append(1)
  else:
    y_data_label.append(2)

x_data, y_data
```

```python
print(y_data_label)
```

## 2. 데이터 시각화
---

* 데이터를 matplotlib.pyplot 라이브러리를 이용하여 시각화 합니다.
  * (예시)

![](https://github.com/yebiny/Lecture_deeplearning_basic/blob/main/imgs/iris_datapresent.png?raw=true)

```python
import matplotlib.pyplot as plt

plt.plot(x_data, y_data, marker='o', linestyle='')
plt.legend( ( "sepal_length", "petal_length", "sepal_width", "petal_width" ) )
plt.show()
```

## 3. 데이터 전처리
---

* y 데이터 값을 원-핫 벡터로 코딩합니다.
* 학습데이터와 테스트데이터를 분리합니다. 
  * sklearn.model_selection에서 제공하는 train_test_split 를 사용합니다.


### **One-hot 코딩**


```python
from tensorflow.keras.utils import to_categorical
y_onehot = to_categorical(y_data_label, num_classes=3)
print(y_onehot) 
```

### **train-test split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data,y_onehot)

print("1:", x_train)
print("2:", x_test)
```

```python
print("1:", y_train)
print("2:", y_test)
```

### **Softmax regression**
